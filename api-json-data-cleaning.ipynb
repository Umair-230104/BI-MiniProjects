```python
import pandas as pd
import requests
```

```python
def show_dataset():
    url = "https://atlasapi.cphmk.dk/api/countries"
    response = requests.get(url)
    data = response.json()
    df = pd.json_normalize(data)
    return df
```

```python
show_dataset()
```
